```python
# üß† Parte 1: A Fundamenta√ß√£o Matem√°tica do CRC

'''O algoritmo trata sequ√™ncias de bits como polin√¥mios com coeficientes bin√°rios (0 ou 1).  
CRC calcula um c√≥digo de verifica√ß√£o com base em uma divis√£o polinomial bin√°ria.

- Mensagem M(x) (m bits)
- Gerador G(x) (r+1 bits, grau r)

### F√≥rmulas:

- Anexa r zeros √† mensagem ‚Üí \( M(x) \cdot 2^r \)
- Divide pelo gerador ‚Üí \( (M(x) \cdot 2^r) \mod G(x) \)
- Resultado (resto) = CRC

Para validar a integridade:
- Se \( T(x) \mod G(x) = 0 \), ent√£o n√£o h√° erro.

### Representa√ß√£o:'''

```

```python
def xor_bits(a, b):
    """
    Realiza a opera√ß√£o de XOR bit a bit entre duas strings bin√°rias de mesmo comprimento.
    """
    resultado = ""
    for i in range(len(a)):
        if a[i] == b[i]:
            resultado += '0'
        else:
            resultado += '1'
    return resultado


def calcular_crc_manual(dados_bits: str, gerador_bits: str) -> str:
    """
    Calcula o CRC para uma sequ√™ncia de dados M(x) usando um gerador G(x).
    
    Args:
        dados_bits: A string bin√°ria representando o polin√¥mio da mensagem, M(x).
        gerador_bits: A string bin√°ria representando o polin√¥mio gerador, G(x).
        
    Returns:
        A string bin√°ria de r bits representando o CRC.
    """
    r = len(gerador_bits) - 1  # grau do polin√¥mio gerador
    
    mensagem_aumentada = list(dados_bits + '0' * r)

    for i in range(len(dados_bits)):
        if mensagem_aumentada[i] == '1':
            janela_atual = "".join(mensagem_aumentada[i : i + r + 1])
            resultado_xor = xor_bits(janela_atual, gerador_bits)
            
            for j in range(len(resultado_xor)):
                mensagem_aumentada[i + j] = resultado_xor[j]

    resto = "".join(mensagem_aumentada[-r:])
    return resto


dados_teste =  "1101011111" 
gerador_teste =  "10011"

crc_calculado = calcular_crc_manual(dados_teste, gerador_teste)

print(f"Dados M(x): {dados_teste}")
print(f"Gerador G(x): {gerador_teste}")
print(f"CRC Calculado: {crc_calculado}")
```

```python
## Parte 3: 

import time
import tracemalloc
import os
import matplotlib.pyplot as plt
from crc import Calculator, Crc16


def xor_bits(a, b):
    resultado = ""
    for i in range(len(a)):
        if a[i] == b[i]:
            resultado += '0'
        else:
            resultado += '1'
    return resultado

def calcular_crc_manual(dados_bits: str, gerador_bits: str) -> str:
    r = len(gerador_bits) - 1
    mensagem_aumentada = list(dados_bits + '0' * r)

    for i in range(len(dados_bits)):
        if mensagem_aumentada[i] == '1':
            janela = "".join(mensagem_aumentada[i : i + r + 1])
            resultado_xor = xor_bits(janela, gerador_bits)
            for j in range(len(resultado_xor)):
                mensagem_aumentada[i + j + 1] = resultado_xor[j]

    resto = "".join(mensagem_aumentada[-r:])
    return resto
```

```python
# Fun√ß√£o para o c√≥digo funcionar

def xor_bits(a, b):
    """
    Realiza a opera√ß√£o de XOR bit a bit entre duas strings bin√°rias de mesmo comprimento.
    """
    resultado = ""
    for i in range(len(a)):
        if a[i] == b[i]:
            resultado += '0'
        else:
            resultado += '1'
    return resultado


def calcular_crc_manual(dados_bits: str, gerador_bits: str) -> str:
    """
    Calcula o CRC para uma sequ√™ncia de dados M(x) usando um gerador G(x).
    
    Args:
        dados_bits: A string bin√°ria representando o polin√¥mio da mensagem, M(x).
        gerador_bits: A string bin√°ria representando o polin√¥mio gerador, G(x).
        
    Returns:
        A string bin√°ria de r bits representando o CRC.
    """
    r = len(gerador_bits) - 1
    mensagem_aumentada = list(dados_bits + '0' * r)
    
    for i in range(len(dados_bits)):
        if mensagem_aumentada[i] == '1':
            janela_atual = "".join(mensagem_aumentada[i : i + r + 1])
            resultado_xor = xor_bits(janela_atual, gerador_bits)
            
            for j in range(len(resultado_xor)):
                mensagem_aumentada[i + j] = resultado_xor[j]
    
    resto = "".join(mensagem_aumentada[-r:])
    return resto
```

```python
# Usando o CRC-16/MODBUS da biblioteca
calculator_lib = Calculator(Crc16.MODBUS)

# Tamanhos em bytes
tamanhos_bytes = [1500, 4500, 9000]

resultados = []

for tamanho in tamanhos_bytes:
    print(f"Analisando mensagem de {tamanho} bytes...")

    mensagem_bytes = os.urandom(tamanho)
    mensagem_bits = "".join(format(byte, '08b') for byte in mensagem_bytes)

    # Manual
    tracemalloc.start()
    start_time = time.perf_counter()

    crc_manual = calcular_crc_manual(mensagem_bits, "11000000000000101") # CRC-16/MODBUS

    end_time = time.perf_counter()
    mem_atual_manual, mem_pico_manual = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    tempo_manual = end_time - start_time

    # Biblioteca
    tracemalloc.start()
    start_time = time.perf_counter()

    crc_lib = calculator_lib.checksum(mensagem_bytes)

    end_time = time.perf_counter()
    mem_atual_lib, mem_pico_lib = tracemalloc.get_traced_memory()
    tracemalloc.stop()

    tempo_lib = end_time - start_time

    resultados.append({
        "tamanho": tamanho,
        "tempo_manual": tempo_manual,
        "mem_pico_manual": mem_pico_manual / 1024,
        "tempo_lib": tempo_lib,
        "mem_pico_lib": mem_pico_lib / 1024
    })

print("Resultados:")
for r in resultados:
    print(r)
```

```python
# Organizar os dados
tamanhos = [r['tamanho'] for r in resultados]
tempos_manual = [r['tempo_manual'] for r in resultados]
tempos_lib = [r['tempo_lib'] for r in resultados]

mem_manual = [r['mem_pico_manual'] for r in resultados]
mem_lib = [r['mem_pico_lib'] for r in resultados]

# Gr√°fico 1: Tempo de Execu√ß√£o
plt.figure(figsize=(10,5))
plt.plot(tamanhos, tempos_manual, marker='o', label='Manual')
plt.plot(tamanhos, tempos_lib, marker='s', label='Biblioteca')
plt.title("Tempo de Execu√ß√£o (Manual vs Biblioteca)")
plt.xlabel("Tamanho da Mensagem (bytes)")
plt.ylabel("Tempo (segundos)")
plt.legend()
plt.grid(True)
plt.show()

# Gr√°fico 2: Pico de Mem√≥ria
plt.figure(figsize=(10,5))
plt.plot(tamanhos, mem_manual, marker='o', label='Manual')
plt.plot(tamanhos, mem_lib, marker='s', label='Biblioteca')
plt.title("Pico de Mem√≥ria (Manual vs Biblioteca)")
plt.xlabel("Tamanho da Mensagem (bytes)")
plt.ylabel("Mem√≥ria (KiB)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import random
from crc import Calculator, Crc16

# Parte 4

# 4.1 - Preparando o cen√°rio
nome = "Maria Clara Rodrigues do Nascimento"
mensagem_base = ''.join(format(ord(c), '08b') for c in nome)

# Polin√¥mio gerador 
gerador_bits = "11000000000000101"  # O ultimo digito da minha matricula √© 1

# Calculando CRC da mensagem base
crc_manual = calcular_crc_manual(mensagem_base, gerador_bits)

# Formando o quadro transmitido (mensagem + CRC)
quadro_transmitido = mensagem_base + crc_manual

# Biblioteca para compara√ß√£o
calculator = Calculator(Crc16.MODBUS)

print(f"Mensagem base: {nome}")
print(f"Mensagem em bin√°rio: {mensagem_base}")
print(f"CRC calculado (manual): {crc_manual}")
print(f"Tamanho do quadro transmitido: {len(quadro_transmitido)} bits")
print("\n----- Testando detec√ß√£o de erros -----\n")

# 4.2 - Rodando os 10 testes de erro
for teste in range(1, 11):
    # Rajada de erro de tamanho aleat√≥rio (entre 1 e 8 bits)
    tamanho_erro = random.randint(1, 8)
    posicao = random.randint(0, len(quadro_transmitido) - tamanho_erro)

    # Gera padr√£o de erro (bits invertidos na posi√ß√£o)
    erro = ['0'] * len(quadro_transmitido)
    for i in range(tamanho_erro):
        erro[posicao + i] = '1'
    erro_bits = ''.join(erro)

    # Quadro corrompido (XOR entre quadro e erro)
    quadro_corrompido = xor_bits(quadro_transmitido, erro_bits)

    # Calcula o CRC do quadro corrompido
    crc_manual_corr = calcular_crc_manual(quadro_corrompido[:-len(crc_manual)], gerador_bits)

    # Biblioteca espera entrada em bytes, convertemos
    quadro_corrompido_bytes = int(quadro_corrompido, 2).to_bytes((len(quadro_corrompido) + 7) // 8, byteorder='big')
    crc_lib_corr = calculator.checksum(quadro_corrompido_bytes)

    # Verifica se os CRCs detectaram erro
    detectou_manual = "Detectou" if crc_manual_corr != '0' * len(crc_manual) else "N√ÉO detectou"
    detectou_lib = "Detectou" if crc_lib_corr != 0 else "N√ÉO detectou"

    # Exibe resultado do teste
    print(f"Teste {teste}:")
    print(f"  - Posi√ß√£o do erro: {posicao}")
    print(f"  - Tamanho da rajada: {tamanho_erro} bits")
    print(f"  - CRC Manual: {detectou_manual}")
    print(f"  - CRC Biblioteca: {detectou_lib}")
    print()
```
